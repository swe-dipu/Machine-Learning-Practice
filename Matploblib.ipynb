```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('Sign Wave')
plt.plot(x,y)
```

```python
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title('Sine')

plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('Cos')
plt.show()
```
